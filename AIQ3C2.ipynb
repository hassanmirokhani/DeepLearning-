```python
import matplotlib.pyplot as plt
```

```python
model = load_model('cats_and_dogs_small.h5')
```

```python
predList = []
```

```python
for i in range(1,10):
    img_path = f'C:/Users/M. Hassan/Downloads/dogs-vs-cats/test/test/{randint(1,1000)}.jpg'
    img = image.load_img(img_path, target_size =(150,150))
    img_tensor = image.img_to_array(img)
    
    img_tensor = np.expand_dims(img_tensor, axis = 0)
    img_tensor /= 255.
    predList.append(model.predict(img_tensor))
   
    for pred in predList:
        if pred>.50:
            str_label = 'Dog'
        else:
            str_label = 'Cat'
    plt.imshow(img_tensor[0])
    plt.title(str_label)
    plt.show()
    
```
